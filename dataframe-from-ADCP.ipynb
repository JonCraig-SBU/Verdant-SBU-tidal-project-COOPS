In this notebook, we are going to make two dataframes for verdant power's ADCP data using Pandas. These dataframes include:
* Depth dependent data
* Depth averaged data

```python
import pandas as pd
import numpy as np
```

First we load each sheet of data for the station that we want as dataframe, then clean the data and come up with the depth-averaged data as time-series, then merge them together.

```python
df1 = pd.read_excel('./14 updated sites/ADCP-Data_Deployment-Period1.xlsx', sheet_name='Station B',skiprows=17)
df2 = pd.read_excel('./14 updated sites/ADCP-Data_Deployment-Period2.xlsx', sheet_name='Station B-2',skiprows=17)
```

## Apply Orientation Corrections

Before merging, we apply orientation corrections to the direction column (column 7: 'Dir') based on analysis from orientation-correction.ipynb:
- **Deployment Period 1**: Dir_corrected = Dir + (-5.53)°
- **Deployment Period 2**: Dir_corrected = Dir + (3.2)°

```python
# Apply orientation correction to Deployment Period 1
# Dir_corrected = Dir + (-5.53)°
correction_p1 = -5.53
if 'Dir' in df1.columns:
    df1['Dir'] = pd.to_numeric(df1['Dir'], errors='coerce') + correction_p1
    print(f"Deployment Period 1: Applied correction of {correction_p1}° to 'Dir' column")
else:
    print("Warning: 'Dir' column not found in df1")
    print(f"Available columns: {list(df1.columns)}")
```

```python
# Apply orientation correction to Deployment Period 2
# Dir_corrected = Dir + (3.2)°
correction_p2 = 3.20
if 'Dir' in df2.columns:
    df2['Dir'] = pd.to_numeric(df2['Dir'], errors='coerce') + correction_p2
    print(f"Deployment Period 2: Applied correction of {correction_p2}° to 'Dir' column")
else:
    print("Warning: 'Dir' column not found in df2")
    print(f"Available columns: {list(df2.columns)}")
```

## Merge Corrected Dataframes
Now that orientation corrections have been applied to both deployment periods, we can merge them together.

```python
df = pd.concat([df1, df2], ignore_index=True)
df.head()
```

```python
columns_to_drop = [c for c in ['Date & Time', 'Date & Time.1', 'Unnamed: 3'] if c in df.columns]
df.drop(columns=columns_to_drop, inplace=True)
```

```python
if 'Unnamed: 8' in df.columns:
    unnamed_8_index = df.columns.get_loc('Unnamed: 8')
    columns_to_drop = df.columns[unnamed_8_index:]
    df.drop(columns=columns_to_drop, inplace=True)
```

```python
df = df.drop(index=[0, 1])
df.head()
```

```python
df.to_csv('depthAvg_ADCPdata.csv', index=False)
```

### Depth dependent data:

```python
tidal_df1 = pd.read_excel('./14 updated sites/ADCP-Data_Deployment-Period1.xlsx', sheet_name='Station B',skiprows=13)
tidal_df2 = pd.read_excel('./14 updated sites/ADCP-Data_Deployment-Period2.xlsx', sheet_name='Station B-2',skiprows=13)
```

```python
# Check what columns are in the depth-dependent dataframes
print("tidal_df1 columns:")
print(list(tidal_df1.columns))
print(f"\nTotal columns: {len(tidal_df1.columns)}")
```

## Apply Orientation Corrections to Depth-Dependent Data
Apply the same orientation corrections to all direction columns in the depth-dependent dataframes:
- **Deployment Period 1**: All Dir columns + (-5.53)°
- **Deployment Period 2**: All Dir columns + (3.2)°

```python
# Apply orientation correction to Deployment Period 1 depth-dependent data
# Dir columns are: 'Unnamed: 7' (depth-averaged) + columns ending in '.3' (depth bins)
correction_p1 = -5.53

# Find all direction columns: 'Unnamed: 7' and columns ending with '.3'
dir_cols_p1 = [col for col in tidal_df1.columns if isinstance(col, str) and col.endswith('.3')]

print(f"Deployment Period 1 - Found {len(dir_cols_p1)} direction columns")
print(f"Direction columns: {dir_cols_p1[:5]}...")  # Show first 5

# Apply correction to all direction columns
for col in dir_cols_p1:
    tidal_df1[col] = pd.to_numeric(tidal_df1[col], errors='coerce') + correction_p1

print(f"Applied correction of {correction_p1}° to all direction columns in tidal_df1")
```

```python
# Apply orientation correction to Deployment Period 2 depth-dependent data
# Dir columns are: 'Unnamed: 7' (depth-averaged) + columns ending in '.3' (depth bins)
correction_p2 = 3.20

# Find all direction columns: 'Unnamed: 7' and columns ending with '.3'
dir_cols_p2 = [col for col in tidal_df2.columns if isinstance(col, str) and col.endswith('.3')]

print(f"Deployment Period 2 - Found {len(dir_cols_p2)} direction columns")
print(f"Direction columns: {dir_cols_p2[:5]}...")  # Show first 5

# Apply correction to all direction columns
for col in dir_cols_p2:
    tidal_df2[col] = pd.to_numeric(tidal_df2[col], errors='coerce') + correction_p2

print(f"Applied correction of {correction_p2}° to all direction columns in tidal_df2")
```

```python
tidal_df = pd.concat([tidal_df1, tidal_df2], ignore_index=True)
tidal_df.head(10)
```

```python
if 'Unnamed: 2' in tidal_df.columns:
    distance_index = tidal_df.columns.get_loc('Unnamed: 2')
    columns_to_drop = tidal_df.columns[:distance_index]
    tidal_df.drop(columns=columns_to_drop, inplace=True)
```

```python
if 'Unnamed: 2' in tidal_df.columns:
    distance_index = tidal_df.columns.get_loc('Unnamed: 2')
    columns_to_drop = tidal_df.columns[:distance_index]
    tidal_df.drop(columns=columns_to_drop, inplace=True)
```
